```python
import traceback
import requests
from bs4 import BeautifulSoup
import csv
from itertools import zip_longest
job_title=[]
job_skill=[]
company_name=[]
location_name=[]
links=[]
salary=[]
page_num=0
desired_page_number=10
counter=0
```

```python
while counter < desired_page_number:
    try:
        url = f"https://wuzzuf.net/search/jobs/?a=navbl&q=data%20engineer&start={page_num}"
        result = requests.get(url)
        src = result.content
        soup = BeautifulSoup(src, "lxml")
        page_limit = int(soup.find("strong").text.replace(",", ""))
        
        if page_num > page_limit // 15:
            print("Page limit reached")
            break
        
        job_titles = soup.find_all("h2", {"class": "css-m604qf"})
        job_skills = soup.find_all("div", {"class": "css-y4udm8"})
        company_names = soup.find_all("a", {"class": "css-17s97q8"})
        locations = soup.find_all("span", {"class": "css-5wys0k"})
        
        for i in range(len(job_titles)):
            # Extract job information and append to respective lists
            job_title.append(job_titles[i].text)
            link = job_titles[i].find("a")
            if link is not None:
                links.append(link['href'] + "https://wuzzuf.net")
            job_skill.append(job_skills[i].text)
            company_name.append(company_names[i].text)
            location_name.append(locations[i].text)
        
        page_num += 1
        counter += 1
        print("Page switched")
    except Exception as e:
        print("Error occurred")
        traceback.print_exc()
        break
```
